# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def urun(ad,fiyat):
    with open("urunler.txt","a", encoding="utf-8") as file:

        file.write(ad + " " +str(fiyat) + "\n" )

urun("iphone 12", 28.000)
urun("iphone 13", 38.000)
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
def degistir(dosyaAdı, eskiKelime, yeniKelime):
    with open(dosyaAdı,"r+") as file:
        text= file.read()
        yeniText= text.replace(eskiKelime, yeniKelime)
        file.seek(0)
        file.write(yeniText)
        file.truncate()

degistir("urunler.txt", "iphone 12", "iphone 11")
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Araç:
    def __init__(self, isim, hız, yakıt):
        self.isim=isim
        self.hız=hız
        self.yakıt=yakıt
    def bilgi(self):
        print(f"isim {self.isim} hız {self.hız} yakıt tuketimi {self.yakıt}")
class Bus(Araç):
    pass
schoolBus=Bus("School Volvo",180,12)
print(f"Vehicle Name= {schoolBus.isim}\n Speed:{schoolBus.hız}\n Mileage: {schoolBus.yakıt}")
````
